Since I want to merge the row that says Nepal along the two datasets, we first extract that row from both the datasets individually and rename that document. un_copy_nepal and un_copy_total. Both documents headers are 'Major area, region, country or area of destination' so you make that equal to Nepal and islolate that row. 

```
#print(UN_Total[UN_Total['Major area, region, country or area of destination']=="Nepal"].transpose().notna())

un_copy_nepal = UN_COPY[UN_COPY['Major area, region, country or area of destination']=="Nepal"]
print(un_copy_nepal)
un_copy_nepal_list = list(un_copy_nepal)

un_copy_total_nepal = UN_Total[UN_Total['Major area, region, country or area of destination']=="Nepal"]
print(un_copy_total_nepal)
un_copy_total_list = list(un_copy_total_nepal)
```

I used un_merge to merge un_copy_nepal and un_copy_total. Those dataframe un_copy_nepal and un_copy_total isolated data entries for Nepal. Use smaller sample size than 8 because i am merging on in one row. 

```
un_merge = un_copy_nepal.merge(un_copy_total_nepal, on= "Major area, region, country or area of destination")                           
print(un_merge.sample(5))
```
